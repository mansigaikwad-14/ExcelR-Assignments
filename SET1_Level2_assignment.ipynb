```python
outcomes = [-2000, -1000, 0, 1000, 2000, 3000]
probab = [0.1, 0.1, 0.2, 0.2, 0.3, 0.1]

most_likely = outcomes[probab.index(max(probab))]

cumulative_probability = sum(prob for outcome, prob in zip(outcomes, probab) if outcome >= 0)

expected_value = sum(outcome * prob for outcome, prob in zip(outcomes, probab))

variance = sum((outcome - expected_value) ** 2 * prob for outcome, prob in zip(outcomes, probab))
standard_deviation = variance ** 0.5

most_likely, cumulative_probability, expected_value, standard_deviation
```
